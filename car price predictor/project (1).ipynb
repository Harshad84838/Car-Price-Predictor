```python
X=df.drop(columns=['selling_price'])
Y=df['selling_price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.15,random_state=14)
```

```python
X.shape,X_train.shape,X_test.shape
```

```python

2232/2627
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import root_mean_squared_error, r2_score, mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
X_train.sample()
```

```python
#linear regression
step1=ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[2,3,4,5,6])],remainder='passthrough')
step2=LinearRegression()
pipe=Pipeline([('step1',step1),('step2',step2)])
pipe.fit(X_train,Y_train)
y_pred=pipe.predict(X_test)
print("R2 Score :   ",r2_score(Y_test,y_pred))
print("RMSE     :   ",root_mean_squared_error(Y_test,y_pred))
print("MAE      :   ",mean_absolute_error(y_pred,Y_test))
```

```python
#lasso
step1=ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[2,3,4,5,6])],remainder='passthrough')
step2=Lasso(alpha=100)
pipe=Pipeline([('step1',step1),('step2',step2)])
pipe.fit(X_train,Y_train)
y_pred=pipe.predict(X_test)
print("R2 Score :   ",r2_score(Y_test,y_pred))
print("RMSE     :   ",root_mean_squared_error(Y_test,y_pred))
print("MAE      :   ",mean_absolute_error(y_pred,Y_test))
```

```python
#RIdge
step1=ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[2,3,4,5,6])],remainder='passthrough')
step2=Ridge(alpha=10)
pipe=Pipeline([('step1',step1),('step2',step2)])
pipe.fit(X_train,Y_train)
y_pred=pipe.predict(X_test)
print("R2 Score :   ",r2_score(Y_test,y_pred))
print("RMSE     :   ",root_mean_squared_error(Y_test,y_pred))
print("MAE      :   ",mean_absolute_error(y_pred,Y_test))
```

```python
#GradientBossting
step1=ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[2,3,4,5,6])],remainder='passthrough')
step2=GradientBoostingRegressor()
pipe=Pipeline([('step1',step1),('step2',step2)])
pipe.fit(X_train,Y_train)
y_pred=pipe.predict(X_test)
print("R2 Score :   ",r2_score(Y_test,y_pred))
print("RMSE     :   ",root_mean_squared_error(Y_test,y_pred))
print("MAE      :   ",mean_absolute_error(y_pred,Y_test))
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import pickle

# 1. Set page config for a neat, centered appearance
st.set_page_config(page_title="Car Price Predictor", page_icon="🚗", layout="centered")

# 2. Cache the heavy files to stop the lagging!
@st.cache_resource
def load_files():
    model = pickle.load(open('pipe.pkl','rb'))
    data = pickle.load(open('df.pkl','rb'))
    return model, data

with st.spinner("Loading Machine Learning Model..."):
    reg_model, df = load_files()

# 3. Main Page Header
st.title("🚗 Car Price Predictor")
st.markdown("Use the menu on the left to enter the car's details and get an instant valuation.")

# 4. Sidebar Inputs (Clean and Native)
st.sidebar.header("📝 Enter Car Details")

year = st.sidebar.slider("Year of manufacture:", min_value=1992, max_value=2020, step=1, value=2015)
km_d = st.sidebar.slider("KM driven:", min_value=100, max_value=200000, step=1000, value=60000)

# We use the EXACT strings that exist in your un-encoded dataframe
fuel_choice = st.sidebar.radio("Fuel Type:", ['Petrol', 'Diesel', 'CNG', 'LPG'], index=0)
transmission_choice = st.sidebar.radio("Transmission:", ['Manual', 'Automatic'], index=0)
owner_choice = st.sidebar.selectbox("Owner:", ['First Owner', 'Second Owner', 'Third Owner', 'Fourth & Above Owner', 'Test Drive Car'], index=0)

brand = st.sidebar.selectbox("Manufacturer:", df['Brand'].unique(), index=1)
model = st.sidebar.selectbox("Model name:", df['Model'].unique(), index=1)

# 5. The Prediction Button
st.divider()
if st.button("💰 Predict Price", use_container_width=True):

    # Bundle the inputs into a Pandas DataFrame so the Pipeline can read the column names
    query_df = pd.DataFrame({
        'year': [year],
        'km_driven': [km_d],
        'fuel': [fuel_choice],
        'transmission': [transmission_choice],
        'owner': [owner_choice],
        'Brand': [brand],
        'Model': [model]
    })

    # Feed the raw text directly into the pipeline!
    op = reg_model.predict(query_df)

    # Show the result neatly
    st.success("Analysis Complete!")
    st.metric(label="Estimated Selling Price", value=f"₹ {int(round(op[0],-2)):,}")
else:
    st.info("👈 Please enter the car details in the sidebar and click Predict.")
```
